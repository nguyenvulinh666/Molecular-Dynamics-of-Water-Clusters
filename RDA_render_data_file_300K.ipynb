```python
distanceOW = 25.2
dis_rOW = [[] for _ in range(4)]

for int in range(len(BigOW)):
    for j in range(20):
        for k in range(j+1, (20)):
            for i in range(np.shape(BigOW[int])[1]):
                x = np.abs(BigOW[int][j][i][0] - BigOW[int][k][i][0])
                x = np.mod(x, distanceOW)

                if abs(x-distanceOW) < x:
                    x = abs(x-distanceOW)

                y = np.abs(BigOW[int][j][i][1] - BigOW[int][k][i][1])
                y = np.mod(y, distanceOW)

                if abs(y-distanceOW) < y:
                    y = abs(y-distanceOW)

                z = np.abs(BigOW[int][j][i][2] - BigOW[int][k][i][2])
                z = np.mod(z, distanceOW)

                if abs(z-distanceOW) < z:
                    z = abs(z-distanceOW)
            
                r = np.sqrt(x**2+y**2+z**2)
                
                # if r <= rcut:
                dis_rOW[int].append(r)
```

```python
counts_dataOW = []
for i in range(4):
    bins_edgeOW = np.linspace(0, 25.196959335562/2*np.sqrt(3), 300)
    countsOW, _ = np.histogram(dis_rOW[i], bins=bins_edgeOW)
    disbin = bins_edgeOW[1] - bins_edgeOW[0]

    countsOW = countsOW /4/np.pi/(bins_edgeOW[:-1]+disbin/2)**2/(disbin)

    normalzieOW = len(dis_rOW[i])

    # print(normalizeOW)
    # plt.plot(bins_edge[1:], counts/normalzie)
    # print(bins_edge)
    counts_dataOW.append(countsOW/normalzieOW*25.196959335562**3)
    # counts_dataOW.append(countsOW)
    a = np.dot(counts_dataOW[i], bins_edgeOW[1:]**2)*4*np.pi*(bins_edgeOW[1] - bins_edgeOW[0]) / 25.196959335562**3
    print(a)
```

```python
# first axis: sub-interval, second axis: number of oxygen, third axis: number of steps, fourth axis: x,y,z.
dip_moment = []

two_nearest_hydrogen = [[], []]
distance = 25.196959335562*np.sqrt(3)
nearest_disance = [distance/2, distance/2] # choose large to find the nearest hydrogen
nearest_dis_index = [0, 0]
# dis = []

for j in range(np.size(modBigOW, axis=1)): # Time Step OW
# for j in range(10): # Time Step OW

    internal_dip = []
    for k in range(np.size(modBigOW, axis=0)): # Oxygen BigOW[i][k][j]
        int_dis = np.zeros(np.size(modBigHW, axis=0))
        for l in range(np.size(modBigHW, axis=0)):
            # print("------------------------" + str(l))
            # internal_dis = np.linalg.norm(modBigOW[k][j] - modBigHW[l][j])
            OH_dis = np.array(modBigOW[k][j] - modBigHW[l][j])

            OH_dis = OH_dis - np.where(OH_dis>0, OH_dis // (25.196959335562/2), OH_dis // (25.196959335562/2) +1) * (25.196959335562/2)
            # print(OH_dis)

            # internal_dis = np.mod(internal_dis, distance)
            # if np.abs(distance - internal_dis) < internal_dis:
            #     internal_dis = np.abs(distance - internal_dis)
            
            int_dis[l] = np.linalg.norm(OH_dis)

            # print("Internal dis", internal_dis)

            # if internal_dis < nearest_distance[np.mod(l,2)]:
            #     nearest_disance[np.mod(l,2)] = internal_dis
            #     nearest_dis_index[np.mod(l,2)] = l
            
        
        mod_int_dis = np.sort(int_dis)[:2]
        
        # print("dis: ", mod_int_dis)

        cut_off = 1.2

        if (mod_int_dis[0] >= cut_off) or (mod_int_dis[1] >= cut_off):
            break
        
        # print(mod_int_dis)

        nearest_dis_index[0] = np.where(int_dis==mod_int_dis[0])[0][0]
        nearest_dis_index[1] = np.where(int_dis==mod_int_dis[1])[0][0]
        # print("------------------------" + str(k))
        
        # print(mod_int_dis)

        # array1 = 2*modBigOW[k][j] - modBigHW[nearest_dis_index[0]][j] - modBigHW[nearest_dis_index[1]][j]
        array1 = modBigOW[k][j] - modBigHW[nearest_dis_index[0]][j]
        array2 = modBigOW[k][j] - modBigHW[nearest_dis_index[1]][j]
        
        array1 = array1 - np.where(array1>0, array1 // (25.196959335562/2), array1 // (25.196959335562/2) +1) * (25.196959335562/2)

        array2 = array2 - np.where(array2>0, array2 // (25.196959335562/2), array2 // (25.196959335562/2) +1) * (25.196959335562/2)

        # print(array1)
        # print(np.linalg.norm(array1))
        # # print(array2)
        # print(np.linalg.norm(array2))
        
        angle = np.arccos(np.dot(array1, array2) / (np.linalg.norm(array1) * np.linalg.norm(array2)))

        # print(angle)
        if angle > 120/180*np.pi:
            break

        big_array = array1 + array2
               
        # big_array = big_array - np.where(big_array>0, big_array // (25.196959335562/2), big_array // (25.196959335562/2) +1) * (25.196959335562/2)

        # Changing unit to debye unit
        transform_unit = 2.01734655502 

        internal_dip.append(((big_array)*transform_unit).tolist())
    dip_moment.append(internal_dip)
    # print("-------------------------")
```
