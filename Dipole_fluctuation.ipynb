```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
diff_temp = [100, 150, 200, 300]
diff_dip_ave = [2.333357276906667, 2.33334535428538, 2.333360578257766,2.333363972354407]
diff_dip_std = [0.00041719834195333964, 0.0005650612213972667, 0.00048640286055455674, 0.0005762107124275893]
```

```python


# Set a modern style
plt.style.use('seaborn-v0_8-whitegrid')  # Clean white background with subtle gridlines
# plt.rcParams['font.family'] = 'Helvetica'  # Professional sans-serif font
plt.rcParams['axes.labelsize'] = 14  # Font size for labels
plt.rcParams['xtick.labelsize'] = 12  # Font size for ticks
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12

# Create the figure and axes
fig, ax = plt.subplots(figsize=(6, 4), dpi=150)  # Modern size and resolution


ax.errorbar(diff_temp, diff_dip_ave, yerr=diff_dip_std, label="Estimates", 
            marker='o', markersize=6, markerfacecolor='green', markeredgecolor='black', 
            markeredgewidth=0.8
            , color='black', ecolor='black', 
            capsize=2, capthick=2, linewidth=1., alpha=0.9)

ax.set_xlabel(r"Temperature [K]", fontsize=14)  # Add x-axis label (adjust based on context)
ax.set_ylabel("Dipole moment [D]", fontsize=14)  # Add y-axis label

ax.grid(True, linestyle='--', alpha=0.5)

# Adjust layout to prevent label clipping
plt.tight_layout()

# Display the plot
# plt.show()
fig.savefig('./figures/Dipole_fluctuation.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
