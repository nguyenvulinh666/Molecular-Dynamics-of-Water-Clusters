# Diffusion constant

```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

OW diffusion constant

```python
dataOW = np.loadtxt("./MDTutorial/Exercises/Exc1/final_exam-nvt-diffusion_constant/displacement-OW.txt")


distanceOW = 25.19
# dis_r = []
BigOW = [[dataOW[(j+1)*800: (j+2)*800][:, i:i+3] for i in range(0, 60,3)] for j in range(4)]
modBigOW = np.concatenate(BigOW, axis=1)
```

```python
len(BigOW[0][0][0])
```

```python
displacement_OW = []

for i in range(1, np.size(modBigOW, axis=1)):
    int_displace = []
    for j in range(np.size(modBigOW, axis=0)):
        int_displace.append(np.linalg.norm(modBigOW[j][i] - modBigOW[j][0])**2)
    displacement_OW.append(int_displace)
```

```python
800*0.0005*100
```

```python
plt.rcParams['axes.labelsize'] = 14  # Font size for labels
plt.rcParams['xtick.labelsize'] = 12  # Font size for ticks
plt.rcParams['ytick.labelsize'] = 12
plt.rcParams['legend.fontsize'] = 12
plt.style.use('seaborn-v0_8-whitegrid')  # Clean white background with subtle gridlines

fig, ax = plt.subplots(figsize=(6, 4), dpi=150)  # Modern size and resolution


timeOW = np.linspace(0, 0.0005*len(displacement_OW)*100*10**-12, len(displacement_OW))

ax.plot(timeOW, np.average(displacement_OW, axis=1)*10**-20, label="Mean square displacement",color="black", alpha=0.8)


slopeOW, interceptOW = np.polyfit(timeOW, np.average(displacement_OW, axis=1)*10**-20, 1)

DOW = slopeOW/6 # diffusion constant m^2/s
print("Diffusion_constant of the oxygen is: ", DOW, "m^2/s")

ax.plot(timeOW, (6*timeOW*DOW), label="Linear-fit", color="orange")
ax.locator_params(axis="y", nbins=4)
ax.locator_params(axis="x", nbins=5)

ax.legend(loc="upper left", fontsize=13)

ax.set_xlabel(r"Time (s)", fontsize=14)  # Add x-axis label (adjust based on context)
ax.set_ylabel(r"$\langle r_{OO}^2 \rangle [\AA^2]$", fontsize=14)  # Add y-axis label
fig.savefig('./figures/Oxygen_diffusion_constant_300K.pdf', format='pdf', dpi=300, bbox_inches='tight')
```
