```python
distanceOW = 25.2
dis_rOW = [[] for _ in range(4)]

for int in range(len(BigOW)):
    for j in range(20):
        for k in range(j+1, (20)):
            for i in range(np.shape(BigOW[int])[1]):
                x = np.abs(BigOW[int][j][i][0] - BigOW[int][k][i][0])
                x = np.mod(x, distanceOW)

                if abs(x-distanceOW) < x:
                    x = abs(x-distanceOW)

                y = np.abs(BigOW[int][j][i][1] - BigOW[int][k][i][1])
                y = np.mod(y, distanceOW)

                if abs(y-distanceOW) < y:
                    y = abs(y-distanceOW)

                z = np.abs(BigOW[int][j][i][2] - BigOW[int][k][i][2])
                z = np.mod(z, distanceOW)

                if abs(z-distanceOW) < z:
                    z = abs(z-distanceOW)
            
                r = np.sqrt(x**2+y**2+z**2)
                
                # if r <= rcut:
                dis_rOW[int].append(r)
```

```python
counts_dataOW = []
bins_edgeOW = np.linspace(2.53781107, 7, 100)
for i in range(4):
    countsOW, _ = np.histogram(dis_rOW[i], bins=bins_edgeOW)
    normalzieOW = np.shape(dis_rOW[i])/(4/3*np.pi*np.max(dis_rOW[i])**3)
    countsOW = countsOW /4/np.pi/bins_edgeOW[1:]**2/(bins_edgeOW[1] - bins_edgeOW[0])
    # plt.plot(bins_edge[1:], counts/normalzie)
    # print(bins_edge)
    counts_dataOW.append(countsOW/normalzieOW)
```

```python
modBigOW = np.concatenate(BigOW, axis=1)
modBigHW = np.concatenate(BigHW, axis=1)
```

```python
# first axis: sub-interval, second axis: number of oxygen, third axis: number of steps, fourth axis: x,y,z.
dip_moment = []

two_nearest_hydrogen = [[], []]
distance = 25.2*np.sqrt(3)
nearest_disance = [distance/2, distance/2] # choose large to find the nearest hydrogen
nearest_dis_index = [0, 0]
# dis = []

for j in range(np.size(modBigOW, axis=1)): # Time Step OW
    internal_dip = []
    for k in range(np.size(modBigOW, axis=0)): # Oxygen BigOW[i][k][j]
        int_dis = np.zeros(np.size(modBigHW, axis=0))
        for l in range(np.size(modBigHW, axis=0)):
            # print("------------------------" + str(l))
            internal_dis = np.linalg.norm(modBigOW[k][j] - modBigHW[l][j])
            internal_dis = np.mod(internal_dis, distance)
            if np.abs(distance - internal_dis) < internal_dis:
                internal_dis = np.abs(distance - internal_dis)
            
            int_dis[l] = internal_dis
            # print(internal_dis)

            # if internal_dis < nearest_distance[np.mod(l,2)]:
            #     nearest_disance[np.mod(l,2)] = internal_dis
            #     nearest_dis_index[np.mod(l,2)] = l
            
        mod_int_dis = np.sort(int_dis)[:2]

        if (mod_int_dis[0] >= 2.5) or (mod_int_dis[1] >= 2.5):
            break
        
        nearest_dis_index[0] = np.where(int_dis==mod_int_dis[0])[0][0]
        nearest_dis_index[1] = np.where(int_dis==mod_int_dis[1])[0][0]
        # print("------------------------" + str(k))
        
        # print(mod_int_dis)

        # array1 = 2*modBigOW[k][j] - modBigHW[nearest_dis_index[0]][j] - modBigHW[nearest_dis_index[1]][j]
        array1 = modBigOW[k][j] - modBigHW[nearest_dis_index[0]][j]
        array2 = modBigOW[k][j] - modBigHW[nearest_dis_index[1]][j]
        
        # print(array1)
        # print(array2)

        array1 = np.mod(array1, distance/np.sqrt(3))
        array1 = np.where(distance/np.sqrt(3)-array1<array1, distance/np.sqrt(3)-array1, array1)

        array2 = np.mod(array2, distance/np.sqrt(3))
        array2 = np.where(distance/np.sqrt(3)-array2<array2, distance/np.sqrt(3)-array2, array2)
        
        # print(array1 + array2)

        internal_dip.append(np.linalg.norm(array1 + array2)*2.0232934131)
    dip_moment.append(list(internal_dip))
```

```python
dip_moment_average = []
dip_moment_std = []

for i in range(len(dip_moment)):
    if dip_moment[i] != []:
        dip_moment_average.append(np.average(dip_moment[i]))

for i in range(len(dip_moment)):
    if dip_moment[i] != []:
        dip_moment_std.append(np.std(dip_moment[i]))
```
